```python
#Target Encoding
from sklearn import base
from sklearn.model_selection import KFold
class KFoldTargetEncoderTrain(base.BaseEstimator,
                               base.TransformerMixin):
    def __init__(self,colnames,targetName,
                  n_fold=5, verbosity=True,
                  discardOriginal_col=False):
        self.colnames = colnames
        self.targetName = targetName
        self.n_fold = n_fold
        self.verbosity = verbosity
        self.discardOriginal_col = discardOriginal_col
    def fit(self, X, y=None):
        return self
    def transform(self,X):
        mean_of_target = X[self.targetName].mean()
        kf = KFold(n_splits = self.n_fold,
                   shuffle=True, random_state=42)
        col_mean_name = self.colnames + '_' + 'Enc'
        X[col_mean_name] = np.nan
        for tr_ind, val_ind in kf.split(X):
            X_tr, X_val = X.iloc[tr_ind], X.iloc[val_ind]
            X.loc[X.index[val_ind], col_mean_name] = X_val[self.colnames].map(X_tr.groupby(self.colnames)
                                     [self.targetName].mean())
            X[col_mean_name].fillna(mean_of_target, inplace = True)
        if self.verbosity:
            encoded_feature = X[col_mean_name].values
            print('Correlation between the new feature, {} and, {} is {}.'.format(col_mean_name,self.targetName,                    
                   np.corrcoef(X[self.targetName].values,
                               encoded_feature)[0][1]))
        if self.discardOriginal_col:
            X = X.drop(self.targetName, axis=1)
        return X
class KFoldTargetEncoderTest(base.BaseEstimator, base.TransformerMixin):
    
    def __init__(self,train,colNames,encodedName):
        
        self.train = train
        self.colNames = colNames
        self.encodedName = encodedName
        
    def fit(self, X, y=None):
        return self
    def transform(self,X):
        mean =  self.train[[self.colNames,
                self.encodedName]].groupby(
                                self.colNames).mean().reset_index() 
        
        dd = {}
        for index, row in mean.iterrows():
            dd[row[self.colNames]] = row[self.encodedName]
        X[self.encodedName] = X[self.colNames]
        X = X.replace({self.encodedName: dd})
        return X
```

```python
trainData = train_data.copy()
trainData = shuffle(trainData)
trainData.reset_index(inplace=True, drop=True)
x = trainData.drop(['booking_bool','click_bool'], axis=1)
y_book = trainData['booking_bool']
y_click = trainData['click_bool']
x_train_book, x_test_book, y_train_book, y_test_book = train_test_split(x, y_book, test_size=.3, random_state=42,shuffle=True)
x_train_click, x_test_click, y_train_click, y_test_click = train_test_split(x, y_click, test_size=.3, random_state=42,shuffle=True)
```

```python
weighting_book = trainData.groupby(['booking_bool']).size()
print(weighting_book)
weighting_click = trainData.groupby(['click_bool']).size()
print(weighting_click)
```

```python
scale_pos_weight_book = weighting_book[0]/weighting_book[1]
print(scale_pos_weight_book)
scale_pos_weight_click = weighting_click[0]/weighting_click[1]
print(scale_pos_weight_click)
```

```python
#make the order back for train data
x_train_book['book'] = y_train_book
x_train_book=x_train_book.sort_values(by=['srch_id'])
y_train_book=x_train_book['book'] 
x_train_book=x_train_book.drop(['book'],axis=1)

#make the order back for test data
x_test_book['book'] = y_test_book
x_test_book=x_test_book.sort_values(by=['srch_id'])
y_test_book=x_test_book['book'] 
x_test_book=x_test_book.drop(['book'],axis=1)

training_group_book=x_train_book.groupby(["srch_id"]).size().values
x_train_book=x_train_book.drop(['srch_id'],axis=1)
test_group_book=x_test_book.groupby(["srch_id"]).size().values
x_test_book=x_test_book.drop(['srch_id'],axis=1)
```

```python
#make the order back for train data
x_train_click['click'] = y_train_click
x_train_click=x_train_click.sort_values(by=['srch_id'])
y_train_click=x_train_click['click'] 
x_train_click=x_train_click.drop(['click'],axis=1)

#make the order back for test data
x_test_click['click'] = y_test_click
x_test_click=x_test_click.sort_values(by=['srch_id'])
y_test_click=x_test_click['click'] 
x_test_click=x_test_click.drop(['click'],axis=1)

training_group_click=x_train_click.groupby(["srch_id"]).size().values
x_train_click=x_train_click.drop(['srch_id'],axis=1)
test_group_click=x_test_click.groupby(["srch_id"]).size().values
x_test_click=x_test_click.drop(['srch_id'],axis=1)
```

```python
# concate Train data, very import here !
xgb_train_book = DMatrix(x_train_book, label = y_train_book)
xgb_train_book.set_group(training_group_book)
# generate eval data
xgb_test_book = DMatrix(x_test_book, label = y_test_book)
xgb_test_book.set_group(test_group_book)

evallist_book = [(xgb_train_book,'train'),(xgb_test_book, 'eval')]
```

```python
params = {
    'bst:max_depth':2, 
    'bst:eta':1, 'silent':1, 
    'objective':'rank:pairwise',
    'nthread':4,
    'eval_metric':'ndcg'
}
```

```python
%time ranking_model_book = train(params, xgb_train_book, evals=evallist_book, num_boost_round=1000, early_stopping_rounds=10)
%time ranking_model_book = train(params, xgb_train_book, evals=evallist_book, num_boost_round=100, xgb_model=ranking_model_book)
```

```python
# concate Train data, very import here !
xgb_train_click = DMatrix(x_train_click, label = y_train_click)
xgb_train_click.set_group(training_group_click)
# generate eval data
xgb_test_click = DMatrix(x_test_click, label = y_test_click)
xgb_test_click.set_group(test_group_click)

evallist_click = [(xgb_train_click,'train'),(xgb_test_click, 'eval')]
params = {
    'bst:max_depth':2, 
    'bst:eta':1, 'silent':1, 
    'objective':'rank:pairwise',
    'nthread':4,
    'eval_metric':'ndcg'
}
%time ranking_model_click = train(params, xgb_train_click, evals=evallist_click, num_boost_round=1000, early_stopping_rounds=10)
%time ranking_model_click = train(params, xgb_train_click, evals=evallist_click, num_boost_round=100, xgb_model=ranking_model_click)
```

```python
import matplotlib.pyplot as plt

xgc_book = ranking_model_book
# saving the feature names to the model
xgc_book.feature_names = x_train_book.columns.to_list()
xgc_click = ranking_model_click
# saving the feature names to the model
xgc_click.feature_names = x_train_click.columns.to_list()
# Create the feature importances plot
fig, ax = plt.subplots(1, 2, figsize=(15,5))
# plot importances with feature weight
xgb.plot_importance(
    booster=xgc_book, 
    importance_type='weight',
    title='Feature Weight',
    show_values=False,
    height=0.5,
    ax=ax[0],
)
xgb.plot_importance(
    booster=xgc_click, 
    importance_type='weight',
    title='Feature Weight',
    show_values=False,
    height=0.5,
    ax=ax[0],
)
```
