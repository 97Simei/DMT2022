## Clean Function

```python
#Change Date time to week label
def process_date_time(data):
    booking = np.array(data['srch_booking_window'])
    date = data['date_time'].copy()
    for i in range(len(date)):
        time = date[i]
        time = (time.split())[0]
        month = int(time[5:7])
        day = int(time[8:10])
        dictMonth = {1:31,2:59,3:90,4:120,5:151,6:181,7:212,8:243,9:273,10:304,11:334,12:365}
        num_week = (dictMonth[month] + day+ int(booking[i]) )
        if num_week>=365:
            num_week = num_week%365
        num_week = num_week//7
        if num_week > 51:
            num_week = num_week%51
        date[i] = num_week
    return date

#Process NaN
def process_nan(data,name):
    if name =='srch_query_affinity_score':
        replace=data['srch_query_affinity_score'].min()
    else:
        replace = 0
    data[name].fillna(replace,inplace=True)
    return data[name]
# def process_nan(data):
#     replace_srch_query_affinity_score=data['srch_query_affinity_score'].min()
#     replace_prop_review_score = 0
#     replace_visitor_hist_starrating=0
#     replace_visitor_hist_adr_usd=0
    
    
#     data['srch_query_affinity_score'].fillna(replace_srch_query_affinity_score,inplace=True)
#     data['prop_review_score'].fillna(replace_srch_query_affinity_score,inplace=True)
#     data['visitor_hist_starrating'].fillna(replace_visitor_hist_starrating,inplace=True)
#     data['visitor_hist_adr_usd'].fillna(replace_visitor_hist_adr_usd,inplace=True)

#     return data
def process_drop(data,droplist):
    columns=data.columns
    for i in droplist:
        if i in columns:
            data = data.drop(i,axis=1)
    return data

def process_comp(data):
    comp=['comp1_rate', 'comp2_rate', 'comp3_rate', 'comp4_rate', 'comp5_rate', 'comp6_rate', 'comp7_rate', 'comp8_rate']
    data[comp]=data[comp].fillna(0)
    data['comp_rate'] = data[comp].mean(axis=1)
   # data = data.drop(['comp1_rate', "comp1_inv", "comp1_rate_percent_diff", 'comp2_rate', "comp2_inv", "comp2_rate_percent_diff", 'comp3_rate', "comp3_inv", "comp3_rate_percent_diff", 'comp4_rate', "comp4_inv", "comp4_rate_percent_diff", 'comp5_rate', "comp5_inv", "comp5_rate_percent_diff", 'comp6_rate', "comp6_inv", "comp6_rate_percent_diff", 'comp7_rate', "comp7_inv", "comp7_rate_percent_diff", 'comp8_rate', "comp8_inv", "comp8_rate_percent_diff"], axis = 1)
    return data['comp_rate']
def process_location_score2(data):
#     df=data[['prop_location_score1','prop_location_score2']]
#     scaler = MinMaxScaler()
#     df = pd.DataFrame(scaler.fit_transform(df), columns = df.columns)
#     imputer = KNNImputer(n_neighbors=1)
#     df = pd.DataFrame(imputer.fit_transform(df),columns = df.columns)
    df=data.groupby('prop_country_id').agg(prop_location_score2_replace=('prop_location_score2','mean')).reset_index()
    data=data.join(df.set_index('prop_country_id'), on='prop_country_id')
    data.prop_location_score2.fillna(data.prop_location_score2_replace,inplace=True)
    data.prop_location_score2.fillna(0,inplace=True)
    return data['prop_location_score2']
    
```

```python
print("********** processing column: date time **********")
train_data['date_time']=process_date_time(train_data)

print("********** processing column: srch_query_affinity_score **********")
train_data['srch_query_affinity_score']=process_nan(train_data,'srch_query_affinity_score')

print("********** processing column: prop_review_score **********")
train_data['prop_review_score']=process_nan(train_data,'prop_review_score')

print("********** processing column: visitor_hist_starrating **********")
train_data['visitor_hist_starrating']=process_nan(train_data,'visitor_hist_starrating')

print("********** processing column: visitor_hist_adr_usd **********")
train_data['visitor_hist_adr_usd']=process_nan(train_data,'visitor_hist_adr_usd')


print("********** processing column: comp_xxx **********")
train_data['comp_rate']=process_comp(train_data)

print("********** processing column: process_location_score2 **********")
train_data['prop_location_score2']=process_location_score2(train_data)
```
